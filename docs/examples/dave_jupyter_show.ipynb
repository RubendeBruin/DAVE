```python
from DAVE.jupyter import *
from DAVE.visual import create_momentline_actors, create_shearline_actors
```

```python
s = Scene()
s.new_axis('beam')
s.new_point('point',parent='beam',position=(10,0,0))
s.new_force('load',parent='point',force=(0,0,-20));
s.update()
```

```python
bma = create_momentline_actors(s['beam'], scale_to=2)

show(s, additional_actors=bma, camera_pos=(5,-30,0), lookat='y', scale=4)
```
