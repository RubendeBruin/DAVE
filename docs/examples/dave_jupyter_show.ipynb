```python
from DAVE.jupyter import *
from DAVE.visual import create_momentline_actors, create_shearline_actors
```

```python
s = Scene()
s.new_axis('beam')
s.new_point('point',parent='beam',position=(10,0,0))
s.new_force('load',parent='point',force=(0,0,-20));
s.update()
```

```python
bma = create_momentline_actors(s['beam'], scale_to=2)

show(s, additional_actors=bma, camera_pos=(5,-30,0), lookat='y', scale=4)
```

```python
bma = create_momentline_actors(s['beam'], scale_to=2)

show(s, additional_actors=bma, camera_pos=(5,-15,0), lookat=(5,0,0))
```

```python
s = Scene()
s.load_scene('res: billy.dave')
s["sb_inner2"].fill_pct = 100
s["sb_outer2"].fill_pct = 100
s["ps_inner2"].fill_pct = 100.0
s["ps_outer2"].fill_pct = 100.0

s["sb_outer7"].fill_pct = 100
s["ps_outer4"].fill_pct = 100

s.solve_statics()

moment_line = create_momentline_actors(s['Billy'], scale_to=25)
shearline = create_shearline_actors(s['Billy'], scale_to=25)
```

```python
show(s, camera_pos = (60.055882766833015, -190.07016017312623, 26.050582829801677),      projection = '2d',     scale = 49.212986584495425,
     lookat = 'y',
     additional_actors=(*moment_line, *shearline),
     painters='X-ray',
     zoom_fit = False)
```

```python
show(s, camera_pos = (60, -200, 25),
     lookat = 'y',
     additional_actors=(*moment_line, *shearline),
     painters='X-ray',
     geometry_scale=0,
     cog_scale=0,
     zoom_fit = True)
```
